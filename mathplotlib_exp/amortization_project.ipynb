```python
# Линейный метод амортизации
import matplotlib.pyplot as plt

def linear_amortization(useful_time, full_cost):
    '''
    int, float → float
    this fuction calculate amortization by using linear method
    '''
    month_time = [month for month in range(1, (useful_time * 12) + 1)]
    depreciation_rate = 1 / (useful_time * 12)
    amort_list = []
    residual_value_list = []
    residual_value = full_cost
    
    for month in month_time:
        amort_list.append(full_cost * depreciation_rate)
        residual_value = residual_value - amort_list[month - 1]
        residual_value_list.append(residual_value)
        
    figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 10))
    ax1.plot(month_time, residual_value_list, color='purple', linewidth=2.5)
    ax1.set_title('Остаточная стоимость. Линейный метод.')
    ax1.set_xlabel('Месяц')
    ax1.set_ylabel('Стоимость')
    ax2.plot(month_time, amort_list, color='red', linewidth=3)
    ax2.set_title('Сумма амортизации. Линейный метод.')
    ax2.set_xlabel('Месяц')
    ax2.set_ylabel('Сумма амортизации')
    
    
        
linear_amortization(10, 2980000)
```

```python
# ускоренный метод
import matplotlib.pyplot as plt

def reduced_remainder_method(useful_time, full_cost, acceleration_factor):
    month_time = [month for month in range(1, (useful_time * 12) + 1)]
    depreciation_rate = 1 / (useful_time * 12)
    amort_list = []
    residual_value_list = []
    residual_value = full_cost
    
    for month in month_time:
        amort_list.append(residual_value * depreciation_rate * acceleration_factor)
        residual_value = residual_value - amort_list[month - 1]
        residual_value_list.append(residual_value)
        
    figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 10))
    ax1.plot(month_time, residual_value_list, color='purple', linewidth=2.5)
    ax1.set_title('Остаточная стоимость. Метод уменьшаемого остатка')
    ax1.set_xlabel('Месяц')
    ax1.set_ylabel('Стоимость')
    ax2.plot(month_time, amort_list, color='red', linewidth=3)
    ax2.set_title('Сумма амортизации. Метод уменьшаемого остатка')
    ax2.set_xlabel('Месяц')
    ax2.set_ylabel('Сумма амортизации')
    
reduced_remainder_method(10, 2980000, 3)
```

```python
# Метод списания стоимости по сумме чисел лет срока полезного использования (кумулятивный)

def cumulative_method(useful_time, full_cost):
    month_time = [month for month in range(1, (useful_time * 12) + 1)]
    time_sum = sum(month_time)
    remaining_time = useful_time * 12
    depreciation_rate = 1 / (useful_time * 12)
    amort_list = []
    residual_value_list = []
    residual_value = full_cost
    
    for month in month_time:
        amort_list.append(full_cost * (remaining_time / time_sum))
        remaining_time += -1
        residual_value = residual_value - amort_list[month - 1]
        residual_value_list.append(residual_value)
    
    figure, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 10))
    ax1.plot(month_time, residual_value_list, color='purple', linewidth=2.5)
    ax1.set_title('Остаточная стоимость. Метод кумулятивный')
    ax1.set_xlabel('Месяц')
    ax1.set_ylabel('Стоимость')
    ax2.plot(month_time, amort_list, color='red', linewidth=3)
    ax2.set_title('Сумма амортизации. Метод кумулятивный')
    ax2.set_xlabel('Месяц')
    ax2.set_ylabel('Сумма амортизации')
    
cumulative_method(10, 2980000)
```
