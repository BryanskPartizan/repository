```python
# number 2
'''
В саду растут деревья, однако берёза растет в среднем на 10% быстрее, чем остальные деревья.

Задача: реализовать с помощью классов модель дерева с использованием логистической функции роста,
модель берёзы реализовать с помощью класса-потомка.

'''
from math import exp

class Tree(object): # создание общего класса деревьев
    def __init__(self, name, age=0, height=0.5, max_height=120000):  # с заданными атрибутами
        self.name = name
        self.age = age
        self.max_height = max_height
        self.start_height = height
        self.height = self.grow_func(age, self.start_height, self.max_height) # определение высоты по функции роста
        
    
    def grow_func(self, age, start_height, max_height=15):  # логистическая функция роста
        b = max_height
        y0 = start_height
        t = age
        result = (b * y0 * exp(t)) / (b + (y0 * (exp(t) - 1)))
        return round(result, 3)
        
    def grow(self):  # метод роста: повышение возраста дерева на 1 год и перерасчет высоты
        self.age += 1
        self.height = self.grow_func(self.age, self.start_height, self.max_height)
    
    def show_info(self):  # метод, отвечающий за раскрытие информации о дереве
        print(
        f'Tree {self.name}, height: {self.height} cm, age: {self.age}')
        
class Birch(Tree): # создание отдельного класса деревьев-берёз, являющегося потомком родителя (Tree)
    def __init__(self, name, age=0, height=0.5, max_height=120000):
        super().__init__(name, age, height, max_height)  # инициализируем конструктор родителя
        self.height = self.grow_func(age, self.start_height, self.max_height)  
             # перегружаем атрибут высоты из-за другой функции роста
        
    def grow_func(self, age, start_height, max_height=15):  # перегружаем метод функции роста с изменёнными параметрами
        b = max_height
        y0 = start_height
        t = age
        result = (b * y0 * exp(1.1 * t)) / (b + (y0 * (exp(1.1 * t) - 1)))
        return round(result, 3)
    

oak = Tree('Oak', 5)  # создаем деревья с одинаковым возрастом
birch = Birch('Birch', 5)

oak.show_info() 
birch.show_info()  # однако высота деревьев - разная. Это - следствие полиморфизма.

oak.grow()
birch.grow()  # рост деревьев на 1 год также показывает разные результаты в росте деревьев

oak.show_info()
birch.show_info()
```
