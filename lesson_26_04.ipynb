```python
import random

'''
Задание 1
1.Подбросьте монету 20 раз
2.подбросьте монету 2000 раз. Сколько раз выпал "герб"?
3.подбросьте монету 20 (100, 10000,100000) раз. Определить долю бросаний в которых выпала "решка"?
4.подбросьте несимметричную монету 20 (100, 10000) раз. Определить долю бросаний в которых выпала "решка"?  
'''

def part_found(list_):
    tails = list_.count('решка')
    tails_part = round((tails / len(list_)) * 100, 2)
    return tails_part
        

list_= ['орёл', 'решка']


first_list = random.choices(list_, k=20)
print(f'===========================ЗАДАНИЕ 1=======================')
print(first_list)

second_list = random.choices(list_, k=2000)
print('\n')
print(f'===========================ЗАДАНИЕ 2=======================')
print(f"Решка выпала {second_list.count('решка')} раз")

third_list = [random.choices(list_, k=20), random.choices(list_, k=100), 
              random.choices(list_, k=10000), random.choices(list_, k=10000)]
print('\n')
print(f'===========================ЗАДАНИЕ 3=======================')
print(f'Подрошено {len(third_list[0])} раз. Доля решки: {part_found(third_list[0])}%')
print(f'Подрошено {len(third_list[1])} раз. Доля решки: {part_found(third_list[1])}%')
print(f'Подрошено {len(third_list[2])} раз. Доля решки: {part_found(third_list[2])}%')
print(f'Подрошено {len(third_list[3])} раз. Доля решки: {part_found(third_list[3])}%')

four_list = [random.choices(list_, weights=[0.4, 0.6], k=20), random.choices(list_, weights=[0.4, 0.6], k=100), 
              random.choices(list_, weights=[0.4, 0.6], k=10000), random.choices(list_, weights=[0.4, 0.6], k=10000)]
print('\n')
print(f'===========================ЗАДАНИЕ 4=======================')
print(f'Подрошено {len(four_list[0])} раз. Доля решки: {part_found(four_list[0])}%')
print(f'Подрошено {len(four_list[1])} раз. Доля решки: {part_found(four_list[1])}%')
print(f'Подрошено {len(four_list[2])} раз. Доля решки: {part_found(four_list[2])}%')
print(f'Подрошено {len(four_list[3])} раз. Доля решки: {part_found(four_list[3])}%')
```

```python
import random
import matplotlib.pyplot as plt

'''
Задача 2. С помощью ГСЧ провести имитацию бросания игрального кубика.
ЗАДАНИЯ
1. постройте гистограмму (столбчатую диаграмму, функция bar()) абсолютных частот
2. построить гистограмму (столбчатую диаграмму) относительных частот
'''

dice = [1, 2, 3, 4, 5, 6]

list_ = random.choices(dice, k=1000000)
absolute_frequency = [list_.count(number) for number in dice]
relative_frequency = [(list_.count(number) / len(list_)) for number in dice]


fig_one, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(15, 10))
ax1.bar(dice, absolute_frequency, color='slateblue', edgecolor='darkblue', linewidth=3)
ax1.set_title('Абсолютная частота')

ax2.bar(dice, relative_frequency, color ='mediumorchid', edgecolor='indigo', linewidth=3)
ax2.set_title('Относительная частота')

print('')
```

```python
import random

'''
Задача 3. С помощью ГСЧ провести имитацию извлечения шаров из урны.
Пусть в урне находятся M белых шаров и N черных. Как случайно вытащить один шар.
Как случайно вытащить два шара (без возврата и с возвратом). Как случайно вытащить K шара (-ов)
(без возврата и с возвратом).
'''

def retriever(balls, number):
    output_balls = random.choices(balls, k=number)
    return output_balls

def retriever_without_refund(balls, number):
    output_balls = random.choices(balls, k=number)
    for ball in output_balls:
        del(balls[balls.index(ball)])
    return output_balls

number = int(input('Введите количество извлекаемых шаров'))
log = input('Возвращать шары в урну: "да" или "нет"? ')
balls = random.choices(['белый', 'черный'], k=1000000)
print(f'В урне {len(balls)} шаров')


if log == 'да':
    result = retriever(balls, number)
else:
    result = retriever_without_refund(balls, number)
    

print(f'Извлечений: {number}.')
print(f'Извлеченные шары: {result}')
print(f'В урне после извлечения: {len(balls)} шаров')
```

```python
import random
import matplotlib.pyplot as plt

'''
Задача 4. Напишите код программы, которая равновероятно разыгрывает следующие варианты: 
red, orange, yellow, green, blue, indigo, violet. Постройте гистограмму (столбчатую диаграмму) цветов радуги.
'''

def root(colors):
    return random.choices(colors, k=100000)

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'indigo', 'violet']
color_data = root(colors)
color_result_data = [color_data.count(color) for color in colors]

fig_one, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(15, 10))
ax1.bar(colors, color_result_data)
ax1.set_title('Частота выбора цвета')
ax1.get_children()[0].set_color('red')
ax1.get_children()[1].set_color('orange')
ax1.get_children()[2].set_color('yellow')
ax1.get_children()[3].set_color('green')
ax1.get_children()[4].set_color('blue')
ax1.get_children()[5].set_color('indigo')
ax1.get_children()[6].set_color('violet')

```

```python
class Player(object):
    def __init__(self, name):
        self.name = name
        
        
class Wallet(Player):
    def __init__(self, name, amount):
        super().__init__(name)
        self.__value = amount
        self._type = self.set_type()
        
    def place_a_bid(self, amount):
        if self.__value - amount >= 0:
            self.__value -= amount
        else:
            swap = amount - self.__value
            self.__value -= (amount - swap)
            self.__value -= self.add_credit(swap)
        self.set_type()
    
    def add_money(self, money):
        self.__value += money
        self.set_type()
        
    def get_value(self):
        return self.__value
    
    def add_credit(self, swap):
        credit_amount = swap * 1.4
        return credit_amount
    
    def set_type(self):
        if self.__value >= 0:
            self.type = 'Не должник'
        else:
            self.type = 'Должник'
    
```
