```python
#1

import matplotlib.pyplot as plt
import numpy as np

a = 0.75
b = 2

lag=0.02
x=np.arange(a,b,lag)
y=np.log(x)-1
z=np.exp(np.cos(x))
fig=plt.figure() 
plt.plot(x, y, label='y=log(x)-1/x', color = 'red', linewidth=2)
plt.plot(x, z, label='exp(cos(x))', color = 'blue', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function: y=log(x)-1/x, function: y=exp(cos(x))')
plt.grid(True)
plt.legend(loc='upper left')
name = 'pic_1'
fmt = 'png'
plt.savefig('{}.{}'.format(name, fmt), fmt='png')
plt.show()
```

```python
#2

import matplotlib.pyplot as plt
import numpy as np # импортирование модулей

x = np.random.randint(low=1, high=11, size=100) # создание массива с целыми числами от 1 до 11 длиной 100
y = x + np.random.randint(1, 5, size=x.size)+np.random.randint(1, 5, size=x.size) # cоздание одномерного массива с числами, отличающимися от массива х
data = np.column_stack((x, y)) # склеивание двух одномерных массивов в один двумерный

fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2,figsize=(8, 4)) #с оздание двух "окон" для вывода туда данных
ax1.scatter(x=x, y=y, marker='o', c='r',edgecolor='black') # создание "кружочков" в форме букв "о" с красным цветом и черной обводкой
ax1.set_title('Поле рассеяния: $x$ и $y$') # подпись для первого "окошка"
ax1.set_xlabel('$x$') # подпись направления "х" для первого окна в виде значений х
ax1.set_ylabel('$y$') # подпись направления "у" для первого окна в виде значений y

ax2.hist(data, bins=np.arange(data.min(), data.max()),label=('x', 'y')) # создание гистограммы из данных двумерного массива в радиусе от минимума и максимума в массиве
ax2.legend(loc=(0.05, 0.8)) # обозначение-легенда х и у в локации 0.05 и 0.8 от левого верхнего угла
ax2.set_title('Частота для $x$ и $y$') # подпись для второго "окошка"
ax2.yaxis.tick_right() # перемещение меток оси "у" справа
```

```python
#3
import matplotlib.pyplot as plt
import numpy as np

def y(x):
    result = 0
    if x > 3.8:
        result = 5 * x * np.log(abs(1 + x))
    if 2.8 <= x <= 3.8:
        result = 10000 * np.exp(-x**2 + 2) # домножено на 10000!!!!
    if x < 2.8:
        result = -np.sin(x)
    return result

x = np.arange(-1, 15, 0.02)
y = [y(x0) for x0 in x]

fig=plt.figure(figsize=(15, 20)) 
plt.plot(x, y, label='f(x)', color = 'blue', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('f(x)')
plt.grid(True)
plt.legend(loc='upper left')

plt.show()

```
