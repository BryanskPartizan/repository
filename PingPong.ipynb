```python
import random
import math

class Player(object):
    def __init__(self, name : str, skill_strike : float, skill_defend : float):
        self.name = name
        self.skill_strike = skill_strike
        self.skill_defend = skill_defend
        self.amount_win = 0
        self.amount_defeat = 0
        self.point = 0
            
            
class PingPong(object):
    def __init__(self, player_one, player_two):
        self.players = [player_one, player_two]
    
    
    def determinant(self, attacker, defender):
        skill_delta = attacker.skill_strike - defender.skill_defend
        function = (math.exp(skill_delta)) / (1 + (math.exp(skill_delta)))
        return 1 if random.random() < function else 0
    
    
    def battle(self):
        while self.players[0].point != 11 and self.players[1].point != 11:
            attacker = random.choice(self.players)
            if self.players.index(attacker):
                defender = self.players[0]
            else:
                defender = self.players[1]
                
            if self.determinant(attacker, defender):
                attacker.point += 1
                print(f'{attacker.name} get one point')
                attacker, defender = defender, attacker
            else:
                attacker, defender = defender, attacker
                
        else:
            if self.players[0].point == 11:
                self.players[0].amount_win += 1
                self.players[1].amount_defeat += 1
            elif self.players[1].point == 11:
                self.players[1].amount_win += 1
                self.players[0].amount_defeat += 1  
            self.players[0].point = 0
            self.players[1].point = 0
            
            
    def playing(self):
        for i in range(5):
            print(f'Новый раунд')
            self.battle()
            print(f'\n{self.players[0].name} - {self.players[0].amount_win} wins')
            print(f'{self.players[1].name} - {self.players[1].amount_win} wins\n')
            
        if self.players[0].amount_win >= 3:
            print(f'Победитель {self.players[0].name}, {self.players[0].amount_win} wins')
        elif self.players[1].amount_win >= 3:
            print(f'Победитель {self.players[1].name}, {self.players[1].amount_win} wins')
            
            
            
if __name__ == '__main__':
    player_one = Player('Emil', 1, 1.3)
    player_two = Player('Erich', 1.4, 1)
    
    battlefield = PingPong(player_one, player_two)
    battlefield.playing()
                
```
