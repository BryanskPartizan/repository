```python
# 1 ВАРИАНТ 23 метод бисекции

import matplotlib.pyplot as plt
import numpy as np

f = lambda x: x**3 + 5*x -4
z = lambda x: x * 0

x = np.arange(-7, 7, 0.0001)
y = f(x)
f1 = z(x)
fig=plt.figure(figsize=(15, 20)) 
plt.plot(x, y, label='y=log(x)-1/x', color = 'purple', linewidth=2)
plt.plot(x, f1, label='y=0', color = 'red', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function: y = x^3 + 5x - 4')
plt.grid(True)
plt.legend(loc='upper left')

a = -10
b = 10
e = 10 ** (-6)

def root(a, b, e):
    x = (a + b) / 2
    while np.fabs(f(x)) >= e:
        x = (a + b) / 2
        a, b = (a, x) if f(a) * f(x) < 0 else (x, b)
    return (a + b) / 2

print(f'f(x) = x^3 + 5x - 4, f(x) = 0 при x = {root(a, b, e)}')
```

```python
# 2 ВАРИАНТ 23

import matplotlib.pyplot as plt
import numpy as np

f = lambda x: 1 + np.exp(-np.sqrt(x)) - np.log(x)
z = lambda x: x * 0

x = np.arange(10**(-2), 10, 0.001)
y = f(x)
f1 = z(x)
fig=plt.figure(figsize=(10, 10)) 
plt.plot(x, y, label='y=1 + exp(-sqrt(x)) - ln(x)', color = 'purple', linewidth=2)
plt.plot(x, f1, label='y=0', color = 'red', linewidth=2)
plt.xlabel('x')
plt.ylabel('y')
plt.title('function: y = 1 + exp(-sqrt(x)) - ln(x)')
plt.grid(True)
plt.legend(loc='upper right')

x0 = 1
x1 = 5
e = 10 ** (-6)


def secant_method(x0, x1, e):
    fx0 = f(x0)
    fx1 = f(x1)

    for _ in range(100):
        if np.fabs(fx1 - fx0) < e:
            return x1
        x2 = (fx1 * (x0 - x1)) / (fx1 - fx0)
        x0,  x1  = x1,  x2
        fx0, fx1 = fx1, f(x2)

print(secant_method(x0, x1, e))
```
